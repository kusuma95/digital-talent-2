# Analisa Data Negara Tujuan Ekspor Indonesia Tahun 2005-2013

## _Mengimpor pandas dan membaca dataset_

```python
import pandas as pd

data_ekspor = pd.read_csv('processed-nilai-ekspor-negara-tujuan-2005-2013.csv', delimiter = ',', encoding = "ISO-8859-1")
data_ekspor.sample(5).style.format({'nilai_ekspor': "{:.2f} $"})
```

### Membaca info datasheet

```python
data_ekspor.info(5)
```

### 1. Deskripsi Data

Data yang digunakan adalah data yang disediakan oleh institusi Bank Indonesia melalui portal data.go.id yakni __"processed-nilai-ekspor-negara-tujuan-2005-2013.csv."__ Data ini menampilkan nilai ekspor yang dilakukan oleh Indonesia. Nilai ekspor pada datasheet ini dijabarkan per bulan untuk masing-masing Negara Tujuan Ekspor selama 8 (delapan) tahun mulai dari tahun 2005 - 2013. Berikut ini beberapa poin penting terkait data tersebut:
- Terdiri dari 6 kolom yakni tahun, bulan, benua, region, nama_negara dan nilai_ekspor. 
- Terdiri dari rows sebanyak 4428
- Delimiter menggunakan tanda koma (,)

Penjelasan mengenai masing-masing kolom pada Dataset ini:

- tahun: Tahun, dtype= Int64
- bulan: Bulan, dtype= object
- benua: Nama Benua, dtype= object
- region: Nama Region, dtype= object
- nama_negara: Nama Negara tujuan, dtype= object
- nilai_ekspor: Nilai ekspor (dalam Ribu USD), dtype= Float64

### 2. Statistik Sederhana

Menghitung statistik sederhana untuk numerical data (dalam hal ini adalah nilai ekspor)

### 2.1 Statistik Nilai Ekspor secara umum dalam kurun waktu 2005-2013

```python
count_negara = data_ekspor['nama_negara'].value_counts().count()
avg_ekspor = data_ekspor['nilai_ekspor'].mean()
max_ekspor = data_ekspor['nilai_ekspor'].max()
min_ekspor = data_ekspor['nilai_ekspor'].min()
tujuan_ekspor_utama = data_ekspor[['nama_negara', 'nilai_ekspor']].groupby('nama_negara').sum()


print('Jumlah Negara Tujuan Ekspor : ', count_negara, 'Negara')
print('\n===============================================')
print('Ekspor per bulan ke masing-masing Negara Tujuan')
print('===============================================')
print('Rata-Rata                   : ', avg_ekspor, '(Ribu USD)')
print('Minimum                     : ', min_ekspor, '(Ribu USD)')
print('Maksimum                    : ', max_ekspor, '(Ribu USD)')

```

### 2.2 Statistik Nilai Ekspor per benua dalam kurun waktu 2005-2013

```python
data_ekspor[['benua','nilai_ekspor']].groupby(['benua']).describe().style.format("{:.2f} $")
```

### 2.3 Statistik Nilai Ekspor per benua dan per tahun

```python
data_ekspor[['benua', 'tahun',  'nilai_ekspor']].groupby(['benua','tahun']).describe().style.format("{:.2f} $")
```
